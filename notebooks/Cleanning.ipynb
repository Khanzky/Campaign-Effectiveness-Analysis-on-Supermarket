# `Latar Belakang`

Perusahaan supermarket merupakan salah satu bisnis ritel yang penting dalam industri perdagangan. Sebagai pemain di pasar yang kompetitif, supermarket bertujuan untuk meningkatkan penjualan guna mempertahankan dan meningkatkan pangsa pasar mereka. Salah satu indikator utama kinerja penjualan adalah total pengeluaran pelanggan (total spent customer), yang mencerminkan seberapa banyak pelanggan menghabiskan uang di supermarket.

Supermarket X sedang menghadapi tantangan dalam meningkatkan penjualan mereka. Mereka mengalami penurunan penjualan tiap tahunnya, yang dapat memiliki dampak negatif pada pertumbuhan perusahaan dan keberlanjutan bisnis. Salah satu pendekatan yang dapat diambil oleh supermarket adalah untuk fokus pada peningkatan total pengeluaran pelanggan. Jika mereka dapat meningkatkan jumlah uang yang dihabiskan oleh setiap pelanggan saat berbelanja, ini dapat berdampak positif pada penjualan keseluruhan. 

# Data Types

**Numerical:**
- Dt_Customer: Date of customer's enrollment with the company
- Year_Birth: Customer's birth year 
- Income: Customer's yearly household income 
- Kidhome: Number of children in customer's household
- Teenhome: Number of teenagers in customer's household
- Recency: Number of days since customer's last purchase 
- MntWines: Amount spent on wine in last 2 years
- MntFruits: Amount spent on fruits in last 2 years
- MntMeatProducts: Amount spent on meat in last 2 years
- MntFishProducts: Amount spent on fish in last 2 years
- MntSweetProducts: Amount spent on sweets in last 2 years
- MntGoldProds: Amount spent on gold in last 2 years
- NumDealsPurchases: Number of purchases made with a discount 
- NumWebPurchases: Number of purchases made through the company’s website
- NumCatalogPurchases: Number of purchases made using a catalog
- NumStorePurchases: Number of purchases made directly in stores
- NumWebVisitsMonth: Number of visits to the company’s website in the last month

**Categorical:**\
a. Ordinal : 
- Education: Customer's education level

b. Nominal : 
- Marital_Status: Customer's marital status 
- Complain: 1 if the customer complained in the last 2 years, 0 otherwise 
- AcceptedCmp1: 1 if the customer accepted the offer in the 1st campaign, 0 otherwise
- AcceptedCmp2: 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise
- AcceptedCmp3: 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise
- AcceptedCmp4: 1 if the customer accepted the offer in the 4th campaign, 0 otherwise
- AcceptedCmp5: 1 if the customer accepted the offer in the 5th campaign, 0 otherwise
- Response: 1 if the customer accepted the offer in the last campaign, 0 otherwise

**Datetime:** 
- Dt_Customer: Date of customer's enrollment with the company

## Handling Outliers

```python
df.describe().T
```

```python
# Mencari outliers
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

- Outlier pada Income

```python
df.Income.describe()
```

```python
outliers = find_outliers_IQR(df['Income'])
print('number of outliers: '+ str(len(outliers)))
print('max outlier value: '+ str(outliers.max()))
print('min outlier value: '+ str(outliers.min()))
outliers
```

Create a new column to categorize the income variable\
`Alasan :`\
Karena memungkinkan jika customer memilki income sebesar outlier tersebut sehingga untuk mengatasi outlier tersebut dengan membuat kolom baru terkait income_category

`Income Categories based on MAD and Median`\
Low Income < 34824 (Median - MAD)\
Middle Income 34824 - 67939 \
Upper Income > 67939 (Median + MAD)

Change the data type of Income to Int

```python
df['Income'] = df['Income'].astype(int)
```

Find MAD 

```python
# find the median and MAD
Median_Income = df.Income.median()
MAD_Income = abs(df.Income - df.Income.median()).median()

Low_Income = Median_Income - MAD_Income
Upper_Income = Median_Income + MAD_Income

Low_Income, Upper_Income
```

```python
# Categorize level of Income
def Income_Category(df):
    if df['Income'] >=  Upper_Income:
        return "Upper Income"
    elif df['Income'] > Low_Income and df['Income'] < Upper_Income:
        return "Middle Income"
    else:
        return "Low Income"
    
    
df['Income_Category'] = df.apply(Income_Category, axis=1)
```

Under 15 years old: Children's age group\
15-24 years old: Young age group\
25-34 years old: Early worker age group\

45-54 years old: Pre-retirement age group\

```python
df.Age.max()
```

```python
def Age_Category(df):
    if df['Age'] < 15 :
        return "Children's"
    elif df['Age'] >= 15 and df['Age'] <= 24:
        return "Young"
    elif df['Age'] >= 25 and df['Age'] <= 34:
        return "Early worker"
    elif df['Age'] >= 35 and df['Age'] <= 44:
        return "Middle-aged"
    elif df['Age'] >= 45 and df['Age'] <= 54:
        return "Pre-retirement"
    elif df['Age'] >= 55 and df['Age'] <= 64:
        return "Retirement"
    else:
        return "Elderly"
    
df['Age_Category'] = df.apply(Age_Category, axis=1)
df
```
